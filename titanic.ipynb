```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")
df
```

```python
age = df["age"].dropna()
age
```

```python
mean_age = age.mean() #avg 
mean_age
```

```python
median_age = age.median() #middle value
median_age
```

```python
mode_age = age.mode()[0]    #most common value
mode_age  
```

```python
range_age = age.max() - age.min() #diff between biggest value and lowest value
range_age
```

```python
variance_population = np.var(age, ddof=0)
variance_population
```

```python
variance_sample = np.var(age, ddof=1)
variance_sample 
```

```python
#passenger ages are quite spread out around the mean
```

```python
std_population = np.std(age, ddof=0)
std_population
#std = square root of variance 
```

```python
std_sample = np.std(age, ddof=1)
std_sample 
```

```python
#he standard deviation (14.5) shows that passenger ages are spread out around the mean
```

```python
# Bar Chart: Number of passengers per class
sns.countplot(x="pclass", data=df)
plt.title("Number of Passengers per Class")
plt.show()
# most passengers where in the 3rd class followed by 1st class, and then 2nd class
```

```python
# Histogram: Age distribution
sns.histplot(df["age"].dropna(), bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()

#Most passengers were aged between 20 and 35, with a noticeable number of children and only a few elderly passengers

```

```python
# Box Plot: Fare by passenger class
sns.boxplot(x="pclass", y="fare", data=df)
plt.title("Fare by Passenger Class")
plt.show()
#First class: Most expensive , has outliers.
#Second & third class: Cheaper, prices more uniform.
#Scatter plot: Shows spread and outliers.
```
