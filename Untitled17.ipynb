# handling nulls and duplications


```python
df.isna().sum()
```

```python
df.drop('deck',axis=1,inplace=True)
```

```python
df
```

```python
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.dropna(subset=['embark_town'], inplace=True)
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
```

```python
Q1,Q3
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['age'] < lower_bound) | (df['age'] > upper_bound)]
```

```python
outliers
```

```python
median_age = df['age'].median()
df['age'].fillna(median_age,inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv("titanic_cleaned.csv")
```

# lab 2 Matplotlib


```python
import matplotlib.pyplot as plt
```

```python
survival_counts = df['survived'].value_counts() 
survival_counts
```

```python
plt.bar(survival_counts.index, survival_counts.values, color=['lightpink', 'lightskyblue'])
plt.xticks([0,1], ['not survived', 'survived'])
plt.xlabel("survival status")
plt.ylabel("number of passengers")
plt.title("survival count")
```

```python
class_counts = df['pclass'].value_counts()
class_counts
```

```python
labels = ['1st class', '2nd class', '3rd class']
plt.pie(class_counts, labels=labels, autopct='%1.1f%%',colors=['lightblue','lightgreen','lightpink'])
plt.title("passenger distribution by classes")
```

```python
embarked_counts = df['embarked'].value_counts()
embarked_counts
```

```python
plt.bar(embarked_counts.index, embarked_counts.values, color=['lightpink', 'lightskyblue', 'lightgreen'])
plt.xlabel("embarked")
plt.ylabel("number of passengers")
plt.title("passenger count by embarkation")
```

```python
colors = df['survived'].map({0: 'lightpink', 1: 'lightskyblue'})
plt.scatter(df["age"], df["fare"],c=colors)
plt.xlabel("age")
plt.ylabel("fare")
plt.title("scatter plot of age vs fare colored by survive")
```

```python
survival_rate = df.groupby('sex')['survived'].mean() * 100
survival_rate
```

```python
plt.bar(survival_rate.index, survival_rate.values, color=['lightpink','lightskyblue'])
plt.xlabel("gender")
plt.ylabel("survival Rate")
plt.title("survival rate by gender")
```
